```python
# Plotting helper

def plot_img_comparison(img_left: np.ndarray, img_right: np.ndarray, fig_title: str) -> None:

    # plot original + processed imgs side by side
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    fig.suptitle(fig_title, fontsize=16)
    
    axes[0].imshow(img_left, cmap='gray', vmin=0, vmax=255)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    axes[1].imshow(img_right, cmap='gray')
    axes[1].set_title('Processed Image')
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
# You may add any helper functions for your implementation. Write your code in the cells below.

# For each call of the function, show on screen the result of your function.

def convolutionMask(img: np.ndarray, mask: np.ndarray, padding: str = 'fill', is_cross_corr: bool = False, display: bool = True, mask_name: str = "N/A") -> np.ndarray:
    """Performs convolution/cross-correlation of image with mask"""

    safe_mask = mask.copy()     # work on copy to protect the global variable

    # flip kernel for cross-correlation
    if is_cross_corr:
        safe_mask = np.flip(safe_mask)

    # run 2D convolution/cross-correlation
    conv_img = convolve2d(
        in1=img, 
        in2=safe_mask, 
        mode="same",
        boundary=padding
    )
    
    # show result on screen
    if display:
        plot_img_comparison(img_left=img, img_right=conv_img, fig_title=f"Convolution with {mask_name}")
    return conv_img
```

### Gaussian Derivative - The Math
* $G_\sigma = c \cdot e^{-\frac{x^2+y^2}{2\sigma^2}} \Longrightarrow$
$\frac{\partial G_\sigma}{\partial x} = \frac{1}{\sigma^2} \cdot (-x) \cdot e^{-\frac{x^2+y^2}{2\sigma^2}} \approx (-x) \cdot G_\sigma $
* Dropping Gaussian coefficient & constant term in derivatve since I'm doing normalization downstream anyway
* Normalization - avoid amplifying/decaying overall pixel intensities
    - derivative kernels sum up to $0$ 
    - regular Gaussian kernels sum up to $1$

```python
# TODO: student implementation

# ---------- Helpers ----------
def create_kernel_coord_grid(s_smooth: float) -> tuple[np.ndarray, np.ndarray]:
    """Helper - creates coordinate grid for Gaussian kernel based on given STD"""

    radius = int(np.ceil(3 * s_smooth))     # radius for window size ~(6*s_smooth + 1)
    y, x = np.mgrid[-radius:radius+1, -radius:radius+1]
    return y, x


def gaussian_2d(x: np.ndarray, y: np.ndarray, sigma: float) -> np.ndarray:
    """Helper - generates 2D Gaussian kernel with given standard deviation"""
    return np.exp(-(x**2 + y**2) / (2 * sigma**2))


def minmax_normalize(arr: np.ndarray) -> np.ndarray:
    """Helper - normalizes array values to [0,1] using min-max normalization"""
    return (arr - np.min(arr)) / (np.max(arr) - np.min(arr) + 1e-8)   # tiny const for safety


def harris_process_plot(img: np.ndarray, ix: np.ndarray, iy: np.ndarray, R: np.ndarray, R_th: np.ndarray, x_coords: np.ndarray, y_coords: np.ndarray) -> np.ndarray:
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    
    # original image
    axes[0, 0].imshow(img, cmap='gray')
    axes[0, 0].set_title('Original Image')
    axes[0, 0].axis('off')
    
    # derivatives Ix,Iy
    axes[0, 1].imshow(ix, cmap='gray')
    axes[0, 1].set_title('Derivative Ix')
    axes[0, 1].axis('off')
    
    axes[0, 2].imshow(iy, cmap='gray')
    axes[0, 2].set_title('Derivative Iy')
    axes[0, 2].axis('off')
    
    # corner strength BEFORE threshold normalized to [0,1]
    R_norm = minmax_normalize(R)
    axes[1, 0].imshow(R_norm, cmap='gray')
    axes[1, 0].set_title('R Before Threshold')
    axes[1, 0].axis('off')
    
    # corner strength AFTER threshold
    R_th_norm = minmax_normalize(R_th)
    axes[1, 1].imshow(R_th_norm, cmap='gray')
    axes[1, 1].set_title('R After Threshold')
    axes[1, 1].axis('off')
    
    # original image annotated withdetected corners
    axes[1, 2].imshow(img, cmap='gray')
    axes[1, 2].scatter(x_coords, y_coords, marker='o', edgecolors='green', facecolors='none', s=20)
    axes[1, 2].set_title('Detected Corners Overlay')
    axes[1, 2].axis('off')
    
    plt.tight_layout()
    
    # convert figure to array (as reqd)
    plots_array = plot2array(fig)
    
    # display in notebook
    plt.show() 
    
    return plots_array


def harris_comparison_plot(img: np.ndarray, experiments: list, fig_title: str = "N/A") -> None:
    """Helper - plots original image with detected corners overlaid (for comparison between different parameter settings)"""

    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    fig.suptitle(fig_title, fontsize=16)

    for i, exp in enumerate(experiments):
        title = exp['title']
        corners = exp['corners']
        axes[i].imshow(img, cmap='gray')
        axes[i].scatter(corners[:, 0], corners[:, 1], marker='o', edgecolors='green', facecolors='none', s=20)
        axes[i].set_title(title)
        axes[i].axis('off')

    plt.tight_layout()
    plt.show()


# ---------- Main Functions ----------
def deriv_gauss_xy(s_smooth: float) -> tuple[np.ndarray, np.ndarray]:
    """Returns x and y derivatives of a 2D Gaussian kernel with STD s_smooth, to be used as filters/masks"""

    # 2D Gaussian
    y, x = create_kernel_coord_grid(s_smooth)
    gauss = gaussian_2d(x=x, y=y, sigma=s_smooth)
    
    # x,y derivatives of the Gaussian (see math above)
    g_dx = -x * gauss 
    g_dy = -y * gauss 
    
    # normalize kernels - sum up to 0
    g_dx = g_dx / np.sum(np.abs(g_dx))
    g_dy = g_dy / np.sum(np.abs(g_dy))
    
    return g_dx, g_dy


def grad_xy(im: np.ndarray, s_smooth: float) -> tuple[np.ndarray, np.ndarray]:
    """Computes x,y gradients of image by convolving with Gaussian derivative kernels"""

    # generate masks
    g_dx, g_dy = deriv_gauss_xy(s_smooth)
    
    # apply masks to image (cross-correlation)
    # padding='symm' to prevent dark edges
    ix = convolutionMask(img=im, mask=g_dx, is_cross_corr=True, padding='symm', display=False)
    iy = convolutionMask(img=im, mask=g_dy, is_cross_corr=True, padding='symm', display=False)
    
    return ix, iy


def harris_corner(im: np.ndarray, s_smooth: float, s_neighb: float, th: float, density_size: int, display: bool, k: float = 0.04) -> tuple[np.ndarray, np.ndarray | None]:
    """
    Finds corners in image using Harris corner detection. 
    :return corners: array of (x,y) corner coordinates 
    :return plots: plots visualizing the steps, encoded as array
    """

    # steps 1-2: Ix, Iy
    ix, iy = grad_xy(im, s_smooth)

    # step 3: Ix^2, Iy^2, Ix*Iy (element-wise)
    ix2 = ix ** 2
    iy2 = iy ** 2
    ixiy = ix * iy

    # step 4 (H matrix entries): 
    # Gaussian smoothing kernel for neighborhood summation
    y, x = create_kernel_coord_grid(s_neighb)
    g_neighb = gaussian_2d(x=x, y=y, sigma=s_neighb)
    g_neighb = g_neighb / np.sum(g_neighb)  # normalize - sum up to 1
    
    # H entries
    h11 = convolutionMask(img=ix2, mask=g_neighb, padding='symm', display=False)
    h22 = convolutionMask(img=iy2, mask=g_neighb, padding='symm', display=False)
    h12 = convolutionMask(img=ixiy, mask=g_neighb, padding='symm', display=False)

    # step 5: compute alt. Harris operator R (option 2)
    det_H = (h11 * h22) - (h12 ** 2)
    trace_H = h11 + h22
    R = det_H - (k * (trace_H ** 2))

    # step 6: threshold R to eliminiate irrelevant corner candidates
    R_threshold = np.where(R > th, R, 0)

    # step 7 (non-max suppression):
    # padding to handle window
    pad1 = (density_size - 1) // 2
    pad2 = (density_size - 1) - pad1
    R_padded = np.pad(
        array=R_threshold, 
        pad_width=((pad1, pad2), (pad1, pad2)),
        mode='constant'
    )
    regions = np.lib.stride_tricks.sliding_window_view(                 # collect (density_size X density_size) image regions (sliding window)
        x=R_padded, 
        window_shape=(density_size, density_size)
    )
    region_maximums = regions.max(axis=(2, 3))                          # find max pixel value in each region
    corner_mask = (R_threshold == region_maximums) & (R_threshold > 0)  # pixel p is corner if: (a) value(p)= region's max AND (b) value(p) > threshold

    # get (x,y) coords of corner (non-zero) pixels
    y_coords, x_coords = np.nonzero(corner_mask)    
    corners = np.column_stack((x_coords, y_coords))

    # display results
    plots = harris_process_plot(img=im, ix=ix, iy=iy, R=R, R_th=R_threshold, x_coords=x_coords, y_coords=y_coords) if display else None

    return corners, plots
```

```python
synthetic_img = np.zeros((600, 800), dtype=np.uint8) # black background

# ROW 1: thickness - solid vs thin outline vs thick outline
cv2.rectangle(synthetic_img, (50, 50), (150, 150), 255, -1)     # solid square
cv2.rectangle(synthetic_img, (300, 50), (400, 150), 255, 1)     # thin outline 
cv2.rectangle(synthetic_img, (550, 50), (650, 150), 255, 15)    # thick Outline


# ROW 2: corner sharpness - sharp vs dull vs curved
# sharp corners (star)
star_pts = np.array([[100, 200], [120, 260], [180, 260], [130, 290], [150, 350], [100, 310], [50, 350], [70, 290], [20, 260], [80, 260]], np.int32)
cv2.fillPoly(synthetic_img, [star_pts], 255)

# dull corners (hexagon)
hex_pts = np.array([[320, 210], [380, 210], [410, 280], [380, 350], [320, 350], [290, 280]], np.int32)
cv2.fillPoly(synthetic_img, [hex_pts], 255)

# no corners (circle)
cv2.circle(synthetic_img, (600, 280), 60, 255, -1)

# ROW 3: edge quality (fuzzy vs jagged/noisy)
# fuzzy/soft edges (with Gaussian blur)
fuzzy = np.zeros_like(synthetic_img)
cv2.rectangle(fuzzy, (50, 420), (150, 520), 255, -1)    # add solid square
fuzzy = cv2.GaussianBlur(fuzzy, (31, 31), 0)            # add blur to soften edges
synthetic_img = cv2.add(synthetic_img, fuzzy)

# jagged/noisy edges (circle with randomized radius)
noisy_circle_pts = []
for angle in np.linspace(0, 2 * np.pi, 80):
    r = 60 + np.random.randint(-2, 2) # add random noise to radius for jaggedness
    x = int(600 + r * np.cos(angle))
    y = int(470 + r * np.sin(angle))
    noisy_circle_pts.append([x, y])
cv2.fillPoly(synthetic_img, [np.array(noisy_circle_pts, np.int32)], 255)


# run test
corners_syn, plots_syn = harris_corner(
    im=synthetic_img, 
    s_smooth=1.0, 
    s_neighb=2.0, 
    th=0.5, 
    density_size=9, 
    display=True
)
```

#### B2.3) Parameter Effects
1. Threshold
    * high threshold makes the algorithm reject weak corners, soft edges, and texture variations where the *gradient* doesn't change drasticlly, only preserving intersections with massive, sharp gradient changes in both the $x$ and $y$ directions
    * seems to focus on corners with high contrast or sharp corners
2. `s_neighb`
    * larger window makes the algorithm use "more context" in detecting gadient changes - more robust to noise like textrures etc., but:
    * it also seems to miss some strong corners or merge 2 adjacent ones
    * larger window seems to introduce some drift between the actual and detected corners
3. `s_smooth`
    * higher `s_smooth` makes the algorith less sensitive to textures and small features, which makes sense due to the blurring effect of a high Gaussian smoothing
    * seems to focus more on macro structures in the image
4. Density Size for Non-Max Supression
    * prevents multiple detections stemming from the same image structure - higher `density_size` means a larger image region inwhich only one detection is allowed
    * with high `density_size`, some structures with multiple visible corners only get a detection on the "strongest" one

##### Failure Case
Some of the corners formed by the scribbles on the left aren't detected. They are the thinner, lighter strokes, and it seems that most of the information about them is lost when computing the R operator. Since the $H$ matrix is computed by taking the determinant (multiplication) of already squared gradient values, it amplifies naturally high-gradient structures like the fabric or origami pieces, while the light scribble pixels dont get inflated as much. There was a tradeoff between filtering out texture noise in the top right part of the image and detecting even the lighter strokes of the scribbles on the left.
